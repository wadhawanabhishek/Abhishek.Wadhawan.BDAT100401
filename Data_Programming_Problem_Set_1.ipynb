```python
#Answering Q3

def triangleArea(a,b,c):
    s = (a+b+c)/2
    ar = (s*(s-a)*(s-b)*(s-c))**0.5
    return ar

area = triangleArea(2,2,2)
print("The area of the triangle is {}".format(area))
```

```python
#Answering Q4
n = int(input("Input the number of elements to be stored in the array "))
li = []
for i in range(n):
    x = int(input("Enter the {} element\n ".format(i)))
    li.append(x)
    
print("\nList is: {}\n".format(li))

li_even = []
li_odd = []
for i in li:
    if i % 2 == 0:
        li_even.append(i)
    else:
        li_odd.append(i)
print("The Even Elements are: {}\n".format(li_even))
print("The Odd Elements are: {}\n".format(li_odd))
```

```python
#Answering Q5 -a

def inside(x,y,x1,y1,x2,y2):
    if (x>x1) & (x<x2) & (y>y1) & (y<y2):
        return True
    else:
        return False

x = inside(1,1,0,0,2,3)
print(x)

y = inside(-1,-1,0,0,2,3)
print(y)
```

```python
#Answering Q6

v = ('a','e','i','o','u')

def pig(s):
    s = s.lower()
    if s.startswith(v):
        return s+'way'
    else:
        new_str = s[1:]
        new_str = new_str+s[0]+'ay'
        return new_str
x = pig('happy')
print(x)

y = pig('Enter')
print(y)
```

```python
name = "bloodtype1.txt"

def bldcount(name):
    infile=open(name,'r')
    lines= infile.readlines()
    for i in lines:
        print(i)
        li=list(i.split(' '))
        li.remove('\n')
        unique_elements=set(li)
        
        for j in unique_elements:
            if li.count(j) > 1:
                print("There are {1} patients of blood type {0}".format(j,li.count(j)))
            else:
                print("There is {1} patient of blood type {0}".format(j,li.count(j)))
                
    infile.close()
            

bldcount(name)
```

```python
#Answering Q8
file_name ='currencies.txt'
dic= {}
with open(file_name) as file:
    for row in file:
        key,value= row[:3],float(row[4:].split(' ')[0].split('\t')[0])
        dic[key]=value
print(dic)
print('\n')
def curcov(curr,conv):
    if curr in dic.keys():
        val= dic.get(curr)
        us_amt = conv* val
        return us_amt
    else:
        print("Currency out of scope")

x = curcov('EUR',100)
print(x)

y = curcov('JPY',100)
print(y)
```

```python
#Answering Q10

def frequencies(s):
    d={}
    stop_words=['.',',',' ']
    s = s.lower()
    s = list(s)
    for i in s:
        if i in stop_words:
            s.remove(i)
    s.sort()
    unq_words= set(s)
    for j in unq_words:
        d[j]=s.count(j)
    
    new_d={k:v for k,v in sorted(d.items())}
    return new_d

x = frequencies('The quick red fox got bored and went home.')
print(x)

print("\n")

y = frequencies("apple")
print(y)
```
